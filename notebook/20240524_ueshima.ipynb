```python
import pandas as pd
import numpy as np
import os
```

```python
df_ID = pd.read_csv('../dataset/exp_dataset/medid_1_v8/ID_split_category.csv')
```

```python
list_val_test = df_ID[df_ID['Split_Category'] != 'Train']['ID'].values # TODO Split Category only have Train Val Test, so exclude the Train dataset

list_val_test 
```

```python
type_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
os.makedirs('../dataset/Ueshima_TrainingData', exist_ok=True)
for type in type_list:
    df = pd.read_csv(f'../dataset/subdataset/medid_1_v8_type{type}/experiment.csv')
    df = df[~df['ID'].isin(list_val_test)]
    df.to_csv(f'../dataset/Ueshima_TrainingData/medid_1_v8_{type}_train.csv', index=False)
    print(f'{type} done')
    print(len(df))
```
